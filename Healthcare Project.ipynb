```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read Dataset: 
df = pd.read_csv("C:/Users/ACER/Downloads/indian_diseases_dataset.csv")
```

```python
# Identify Missing values

df.isnull().sum()
```

```python
# 1. comorbidity :
# Replace Comorbidity missing values with "unconfirmed"

df["comorbidity"] = df["comorbidity"].fillna("unconfirmed")
```

```python
df["comorbidity"].isnull().sum()
```

```python
# 2.  alcohol_use:
# Replace alcohol_use missing values with "--"

df["alcohol_use"] = df['alcohol_use'].fillna("--")
```

```python
df["alcohol_use"].isnull().sum()
```

```python
# 3.  cause_of_death:
# Replace cause_of_death missing values with "Not Defined"

df["cause_of_death"] = df["cause_of_death"].fillna("Not Defined")
```

```python
df["cause_of_death"].isnull().sum()
```

```python
# 4.  recovery_days:
# Replace recovery_days missing values with "--"

df["recovery_days"] = df["recovery_days"].fillna("-")
```

```python
df["recovery_days"].isnull().sum()
```

```python
# To check there is any null values

df.isnull().sum()
```

##### Visualization on Age Distribution of Patients:

```python
# Create histogram
counts, bins, patches = plt.hist(df['age'], bins=15, color='#4E79A7', edgecolor='black')

# Add data labels (centered on each bar)
for i in range(len(counts)):
    plt.text((bins[i] + bins[i+1]) / 2, counts[i], int(counts[i]),
             ha='center', va='bottom', fontsize=7, fontweight = 'bold', color='black')


# Titles and labels
plt.title('Age Distribution', color='darkblue', fontsize=12, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

#### Visualization on Gender Distribution:

```python
ax=sns.countplot(x = 'gender', data=df)


# Add data labels on bars:
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black', fontweight= 'bold')

# Title and axis labels:
plt.title("Gender Distribution",  fontsize=12, fontweight='bold', color='darkblue')
plt.xlabel("Gender", fontsize= 9, fontweight = 'bold', color='black')
plt.ylabel("Count" , fontsize= 9, fontweight = 'bold', color='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
